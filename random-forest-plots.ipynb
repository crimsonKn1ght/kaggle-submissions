```python
weight_bins = [30, 65, 100, 130, float('inf')]
age_bins = [10, 19, 25, 35, float('inf')]
height_bins = [1.4, 1.58, 1.67, 1.74, 1.78, 1.9, float('inf')]
NCP_bins = [0, 2, 3.5, float('inf')]
FCVC_bins = [0, 1.5, 2.5, float('inf')]
CH2O_bins = [0, 1.5, 2.5, float('inf')]
FAF_bins = [0, 0.5, 1.5, 2.5, float('inf')]

train_data['Weight_range'] = pd.cut(train_data['Weight'], bins = weight_bins)
train_data['Height_range'] = pd.cut(train_data['Height'], bins = height_bins)
train_data['Age_range'] = pd.cut(train_data['Age'], bins = age_bins)
train_data['NCP_range'] = pd.cut(train_data['NCP'], bins = NCP_bins)
train_data['FCVC_range'] = pd.cut(train_data['FCVC'], bins = FCVC_bins)
train_data['CH2O_range'] = pd.cut(train_data['CH2O'], bins = CH2O_bins)
train_data['FAF_range'] = pd.cut(train_data['FAF'], bins = FAF_bins)

test_data['Weight_range'] = pd.cut(test_data['Weight'], bins = weight_bins)
test_data['Height_range'] = pd.cut(test_data['Height'], bins = height_bins)
test_data['Age_range'] = pd.cut(test_data['Age'], bins = age_bins)
test_data['NCP_range'] = pd.cut(test_data['NCP'], bins = NCP_bins)
test_data['FCVC_range'] = pd.cut(test_data['FCVC'], bins = FCVC_bins)
test_data['CH2O_range'] = pd.cut(test_data['CH2O'], bins = CH2O_bins)
test_data['FAF_range'] = pd.cut(test_data['FAF'], bins = FAF_bins)
```

```python
model = RandomForestClassifier(n_estimators=175, max_depth=13, random_state=1)

# no change in efficiency of results when estimators set to 150 or 175
# max depth efficiency capped at 13 for now
```

```python
Y_train = train_data['NObeyesdad']
features = ['Gender', 'Age_range', 'Height_range', 'Weight_range', 'family_history_with_overweight', 'FCVC', 'SCC', 'NCP_range', 'FAF_range', 'CAEC', 'CH2O_range', 'FAF']
# need to check later by reducing features

train_data = pd.get_dummies(train_data[features])
test1_data = pd.get_dummies(test_data[features])

model.fit(train_data, Y_train)

predictions = model.predict(test1_data)

output = pd.DataFrame({'id': test_data.id, 'NObeyesdad': predictions})
output.to_csv('submission.csv', index=False)
print("Done!")
```
